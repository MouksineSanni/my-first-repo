```python
# WE HAVE TO MAKE GC COUNTER
#WE HAVE a DNA/ Genome in which we have to count the number of G and C content
DNA=["A", "T", "G", "C", "G", "A", "C", "A", "A","C","T","T"]
#this is the basic value on which we have to perform program
#this is the first input
print(len(DNA))
g_count=0
t_count=0
a_count=0
c_count=0 # Initialize c_count

#start counting the list and count each item which is placed inside of it
#and then place this to the relevant variable
for base in DNA:
  if base=="G":
     g_count= g_count+1 # Corrected variable name and assignment operator
  elif base=="C": # Added colon
     c_count = c_count+1
  elif base=="T": # Added colon
     t_count = t_count+1
  elif base=="A": # Added colon
     a_count = a_count+1
#4.The calculations (Combining Data)
     gc_content=(g_count+c_count)
     total_count=g_count+c_count+t_count+a_count
# The final Lab Report
print("DNA Composition Report") # Added quotes
print(f"GU count: {g_count}")
print(f"CY count: {c_count}")
print(f"TY count: {t_count}")
print(f"AD count: {a_count}")
print(f"GC content: {gc_content}")
print(f"Total count: {total_count}")
```

```python
#1.The Genetic Code Dictionnary
genetic_code = { # Changed to dictionary (curly braces)
    "ATA":"Isoleucine", "ATC":"Isoleucine", "ATT":"Isoleucine", "ATG":"Methionine",
    "ACA":"Threonine", "ACC":"Threonine", "ACG":"Threonine", "ACT":"Threonine", # Fixed duplicate key: changed second "ACA" to "ACC"
    "AAC":"Asparagine", "AAT":"Asparagine", "AAA":"Lysine", "AAG":"Lysine", # Fixed missing value for AAT
    "AGC":"Serine", "AGT":"Serine", "AGA":"Arginine", "AGG":"Arginine",
    "CTA":"Leucine", "CTC":"Leucine", "CTG":"Leucine", "CTT":"Leucine", # Fixed missing quote for CTT
    "CCA":"Proline", "CCC":"Proline", "CCG":"Proline", "CCT":"Proline", # Added missing comma
    "CAC":"Histidine", "CAT":"Histidine", "CAA":"Glutamine", "CAG":"Glutamine",
    "CGA":"Arginine", "CGC":"Arginine", "CGG":"Arginine", "CGT":"Arginine",
    "TCA":"Serine", "TCC":"Serine", "TCG":"Serine", "TCT":"Serine", # Added missing comma
    "TTC":"Phenylalanine", "TTT":"Phenylalanine", "TTA":"Leucine", "TTG":"Leucine", # Added missing comma
    "TAC":"Tyrosine", "TAT":"Tyrosine", "TAA":"Stop", "TAG":"Stop", "TGA":"Stop", # Added missing comma
    "GGA":"Glycine", "GGC":"Glycine", "GGG":"Glycine", "GGT":"Glycine", # Added missing comma
    "GCA":"Alanine", "GCC":"Alanine", "GCG":"Alanine", "GCT":"Alanine", # Added missing comma
    "GAC":"Aspartic Acid", "GAT":"Aspartic Acid", "GAA":"Glutamic Acid", "GAG":"Glutamic Acid",
    "GTA":"Valine", "GTC":"Valine", "GTG":"Valine", "GTT":"Valine", # Added missing comma
    "TGC":"Cysteine", "TGT":"Cysteine", "TGG":"Tryptophan" # Fixed unterminated string literal ("Cysteine,") and removed trailing comma before closing brace
} # Changed to dictionary (curly braces)

#2.Our Input DNA Sequence (The sample)
dna_sequence ="ATGGCCATTGTAATGGGCCGCTGAAAGGGT"

#3.An Empty List to hold the resulting protein sequence
protein_sequence = [] # Initialized an empty list to store the protein sequence

#4.The Loop:Step through the DNA 3 bases at a time
for i in range(0, len(dna_sequence), 3): # Corrected dna_seq to dna_sequence
    #Slice the DNA into a "Codon"
    codon = dna_sequence[i:i+3]
    # Check if the codon is valid (3 bases long) and exists in the genetic code
    if len(codon) == 3 and codon in genetic_code:
        amino_acid = genetic_code[codon]
        protein_sequence.append(amino_acid)
    else:
        # Handle cases where codon is incomplete or not found
        if len(codon) < 3:
            print(f"Warning: Incomplete codon '{codon}' at end of sequence.")
        else:
            print(f"Warning: Unknown codon '{codon}'. Skipping.")

# Optional: print the resulting protein sequence
print(f"DNA Sequence: {dna_sequence}")
print(f"Protein Sequence: {protein_sequence}")
```
