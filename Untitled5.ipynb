```python
# WE HAVE TO MAKE GC COUNTER
#WE HAVE a DNA/ Genome in which we have to count the number of G and C content
DNA=["A", "T", "G", "C", "G", "A", "C", "A", "A","C","T","T"]
#this is the basic value on which we have to perform program
#this is the first input
print(len(DNA))
g_count=0
t_count=0
a_count=0
c_count=0 # Initialize c_count

#start counting the list and count each item which is placed inside of it
#and then place this to the relevant variable
for base in DNA:
  if base=="G":
     g_count= g_count+1 # Corrected variable name and assignment operator
  elif base=="C": # Added colon
     c_count = c_count+1
  elif base=="T": # Added colon
     t_count = t_count+1
  elif base=="A": # Added colon
     a_count = a_count+1
#4.The calculations (Combining Data)
     gc_content=(g_count+c_count)
     total_count=g_count+c_count+t_count+a_count
# The final Lab Report
print("DNA Composition Report") # Added quotes
print(f"GU count: {g_count}")
print(f"CY count: {c_count}")
print(f"TY count: {t_count}")
print(f"AD count: {a_count}")
print(f"GC content: {gc_content}")
print(f"Total count: {total_count}")
```
