```python
#Count GC nucleotides using a loop
sequence = "ATCCGGCTATGCCAAGCTAGCGGTA"
gc_count = 0  #accumulator - starts at zero
for nucleotide in sequence:
    if nucleotide  == "G" or nucleotide == "c":
      gc_count = gc_count + 1 #or: gc_count += 1
gc_content = (gc_count / len(sequence))*100
print(f"GC count: [gc_count]")
print(f"GC content: [gc_content:.if]%")
#Simulating population growth
population = 100
growth_rate = 0.30 # 30% per generation
print("Population over 8 generations:")
for gen in range(1, 9):
    population = population * (1+growth_rate)
    print(f" Generation [gen]: [population:.of] individuals")
```
